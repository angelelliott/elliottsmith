Before I can start any analysis, I must manually gather the Spotify URI of each artist and save it into a dictionary of favorites.

```python
artists = {
    "Elliott Smith":"2ApaG60P4r0yhBoDCGD8YG",
    "Anitta":"7FNnA9vBm6EKceENgCGRMb",
    "Childish Gambino":"73sIBHcqh3Z3NyqHKZ7FOL",
    "Chlöe":"1FtBEIWAwvw5ymBen5GICR",
    "Kurt Vile":"5gspAQIAH8nJUrMYgXjCJ2",
    "Kurt Cobain":"6pAuTi6FXi6qFQJ1dzMXQs",
    "Radiohead":"4Z8W4fKeB5YxbusRsdQVPb",
    "Kendrick Lamar":"2YZyLoL8N0Wb9xBt1NhZWg",
    "Nirvana":"6olE6TJLqED3rqDCT0FyPh",
    "Kehlani":"0cGUm45nv7Z6M6qdXYQGTX",
    "Courtney Barnett":"4OOlG5eBXSkSAAEeKjJb5Y",
    "Hole":"5SHQUMAmEK5KmuSb0aDvsn",
    "Weyes Blood":"3Uqu1mEdkUJxPe7s31n1M9",
    "Mitski":"2uYWxilOVlUdk4oV9DvwqK"
    }

for name, uri in artists.items():
  artists[name] = "spotify:artist:" + uri
```

```python
genres = {}
popularity = {}
```

```python
for artist in spotify.artists(artists.values())['artists']:
  genres[artist['uri']] = artist['genres']
  popularity[artist['uri']] = artist['popularity']
```

```python
artists_df = pd.DataFrame({'name':artists.keys(), 'genres': genres.values(), 'popularity': popularity.values(), 'artisturi': artists.values()})
artists_df.head()
```

```python
albums_df = pd.DataFrame({'name':{},
                          'artist': {},
                          'release_date': {},
                          'total_tracks': {},
                          'albumuri': {},
                          'artisturi':{}})
albums_df.head()
```

```python
for artisturi in artists_df['artisturi']:
  for album in spotify.artist_albums(artisturi, album_type="album")['items']:
    temp = pd.DataFrame({'name': [album['name']],
                         'artist': [album['artists'][0]['name']],
                         'release_date': [album['release_date']],
                         'total_tracks': [album['total_tracks']],
                         'albumuri': [album['uri']],
                         'artisturi': [artisturi]}
                    )
    albums_df = pd.concat([albums_df, temp], ignore_index=True)
```

```python
albums_df.head()
```
